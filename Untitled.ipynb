```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
print(f"The number of school is:  {len(school_data)}")
student_data = pd.read_csv(student_data_to_load)
print(f"The number of students is:  {len(student_data)}")
total_budget = school_data["budget"].sum()
print(f"The total Budget is:  $ {total_budget}")
average_math = student_data["math_score"].mean()
print(f"The average math score is:  {average_math}")
average_reading = student_data["reading_score"].mean()
print(f"The average reading score is:  {average_reading}")
passing_rate = (average_math+average_reading)/2
print(f"The overall passing rate is:  {passing_rate}")
# school_data
# student_data
# school_data_complete
```

```python
# Student overall score
school_data_complete["student overall score"] = (school_data_complete["math_score"]+school_data_complete["reading_score"])/2
school_data_complete
```

```python
# To sort from lowest to highest, ascending=False must be passed in
syudent_overall_sorting = school_data_complete.sort_values("student overall score", ascending=True)
syudent_overall_sorting.head(5)
```

```python
# Adding overall passing to the data frame
school_data_complete["Overall Passing per Student"] = (average_math+average_reading)/2
school_data_complete
```
